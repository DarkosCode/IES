1) Crear una función que reciba un número como parámetro y devuelva True si es primo y False si no lo es

```python
import math

def esPrimo(num):
    if num == 2:
        return True
    if num < 1 or num % 2 == 0:
        return False
    
    limite = math.isqrt(num)
    for i in range(3, limite +1, 2):
        if num % i == 0:
            return False
    return True

print(esPrimo(1))
```

2) Utilizando la función del punto 1, realizar otra función que reciba de parámetro una lista de números y devuelva sólo aquellos que son primos en otra lista

```python
def listaPrimos(lista):
    nuevaLista = []
    for n in lista:
        if esPrimo(n):
            nuevaLista.append(n)
    return nuevaLista

#* Falta agregar que valide que sea una lista de numeros enteros

print(listaPrimos([1,7,9,34,5,7]))
```

3) Crear una función que convierta entre grados C, Farenheit y Kelvin<br>
Fórmula 1	: (°C × 9/5) + 32 = °F<br>
Fórmula 2	: °C + 273.15 = °K<br>
Debe recibir 3 parámetros: el valor, la medida de orígen y la medida de destino


```python
def convertidor(temperatura, origen, destino):
    if origen == "C":
        if destino == "F":
            return ((temperatura * 9/5) + 32, "°F")
        else:
            return (temperatura + 273.15, "°K")
    elif origen == "K":
        if destino == "F":
            return (temperatura - 459.67, "°F")
        else:
            return (temperatura - 273.15, "°C")
    elif origen == "F":
        if destino == "K":
            return ((temperatura - 32) * 5/9 + 273.15, "°K")
        else:
            return ((temperatura - 32) * 5/9, "°C")
    
print(convertidor(32, "K", "F"))

```

5) Armar una función que devuelva el factorial de un número. Tener en cuenta que el usuario puede equivocarse y enviar de parámetro un número no entero o negativo

```python
#* multiplicar dicho número por todos los números enteros positivos inferiores a él hasta llegar a 1
def factorializador(num):
    if type(num) != int:
        print("Debe ser un numero entero!")
    else:
        factorial = num
        multiplicador = num
        while multiplicador > 1:
                multiplicador -= 1
                factorial *= multiplicador
        return factorial
factorializador(9)
```

6) Crear una función que al recibir una lista de números, devuelva el que más se repite y cuántas veces lo hace. Si hay más de un "más repetido", que devuelva cualquiera

```python
listaNumerica = [9,7,7,5,4,6,9,1,2,3,4,7,5]
def contador(listtt):
    mayor = 0
    numerito = 0
    for i in listtt:
        num = listtt.count(i)
        if num > mayor:
            mayor = num
            numerito = i
    return f"El numero {numerito}, {mayor} veces."

print(contador(listaNumerica))
```
